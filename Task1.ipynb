 #          **Task 1: Data Cleaning and Preprocessing**





## **Importing the necessary libraries**




```python
import numpy as np
import pandas as pd
```

## **Reading the dataset**

```python
df = pd.read_csv('/content/BigMart Sales.csv')
print(df)
```

```python
df.info()
```

## **Identify the missing values.**


```python
df.isnull().sum()
```

## **Handling the null values.**





```python
df['Item_Weight'] = df['Item_Weight'].replace(np.nan,df['Item_Weight'].mean())
```

```python
df.fillna(method='ffill',inplace=True)
```

## **After Handling the null values.**

```python
df.isnull().sum()
```

##**Remove duplicate rows.**

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

## **Standardize text values like gender, country names, etc.**


```python
df['Item_Fat_Content'] = df['Item_Fat_Content'].replace({
    'Low Fat': 'LF',
    'Regular': 'Reg'
})
```

```python
print(df)
```

## **Convert date formats to a consistent type (e.g., dd-mm-yyyy).**






```python
df['Date'] = pd.to_datetime('2023-06-01')
```

```python
df
```

```python
df['Date'] = df['Date'].dt.strftime('%d-%m-%Y')
```

```python
df
```

## **Rename column headers to be clean and uniform.**






```python
df.rename(columns={
    'Item_Identifier':'Item_ID'
},inplace=True)
```
